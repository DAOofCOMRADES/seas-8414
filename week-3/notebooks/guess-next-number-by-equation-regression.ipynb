# Import Modules

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

# Generate Data

```python
m = 3
```

```python
d = { "input": list(), "output": list()}
for i in range(100):
    d["input"].append(i)
    d["output"].append(i * m + 5)
pd.
```

```python
dataset = pd.DataFrame(d)
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
dataset[dataset.input % 2 == 0].head(5).tail(3).style.hide_index()
```

# Run Machine Learning Algorithm 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
global regressor
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

# Test the Model

```python
input_value = 200
```

```python
def predict(n):
    global regressor
    return round(regressor.predict([[n]])[0])
```

```python
print("For input", input_value, "prediction is", predict(input_value))
```
