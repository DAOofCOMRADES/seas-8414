# Simple Linear Regression 

```python
x = [2, 3, 5, 7, 9]
y = [4, 5, 7, 10, 15]
```

```python
print("x = ", x)
print("y = ", y)
```

Equation of line that best fits:

## y = mx + b

Reference: https://www.mathsisfun.com/data/least-squares-regression.html



## 1. Calculate x^2

```python
x2 = [i ** 2 for i in x]
x2
```

## 2. Calculate x * y

```python
xy = [x[i] * y[i] for i in range(len(x))]
xy
```

## 3. Sum of x, y, x^2, and xy

```python
sum_x = sum(x) # sum of values of x
sum_y = sum(y) # sum of values of y
sum_x2 = sum(x2) # sum of values of x2
sum_xy = sum(xy)  # sum of values (x * y)
n = len(x) # number of samples for a variable "X"
```


## 4. Calcuate 

m =  ((N * Σ(xy)) − (Σx * Σy)) / ((N * Σ(x2)) − ((Σx)^2)) 


```python
m = ((n * sum_xy) - (sum_x * sum_y)) / ((n * sum_x2) - (sum_x) ** 2)
print("m = ", m)
```

## 5. Calculate b = Σy − m ΣxN

```python
b = (sum_y - m * sum_x) / n
print("b = ", b)
```

```python
x = [i for i in range(20, 25)]
for x1 in x:
    y = m * x1 + b
    print(f"For input {x1} predicting {y}")
```
