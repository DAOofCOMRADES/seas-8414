===================== Import required libraries ======================

```python
# importing required libraries
import pandas as pd
import sklearn.datasets as datasets
from sklearn.preprocessing import OrdinalEncoder
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
```

```python
data = """Name	Age of best work	Art Work	# of Art Pieces	Success
DaVinci	35	35K	3	No
Angelo	22	50K	2	No
Picaso	63	10K	1	Yes
Pollack	59	170K	1	Yes
Raphael	25	40K	4	No
Dr. Mallarapu	37	50K	2	?"""


data = pd.DataFrame([l.split("\t") for l in data.split("\n")])
data.columns = data.iloc[0]
data = data[1:]
data
```

========================== Hyperparameters ===========================

```python
# Set Hyperparameters
k = 2
search_algo = 'auto'
dist = 'euclidean'
```

============================== Data set ===============================

```python
feature_names = data.columns
classes = data["Success"].unique()
```

```python
enc = OrdinalEncoder()
enc.fit(data)
data = pd.DataFrame(enc.transform(data))
data.columns = feature_names
data
```

```python
X = data.drop(["Success"], axis=1)
y = data["Success"]
X_train = X.iloc[:-1]
X_test = X.iloc[-1].to_frame().T
y_train = y[:-1]
```

```python
X_test
```

============================== Training ===============================

```python
# Train k-Nearest Neighbor Model
model = KNeighborsClassifier(n_neighbors=k, algorithm=search_algo, metric=dist)
model.fit(X_train, y_train)
result = model.predict(X_test)
```

```python
if result == 1.0:
    print("Dr. Mallarapu will NOT be successful as Artist")
else:
    print("Dr. Mallarapu will be successful as Artist")
```
